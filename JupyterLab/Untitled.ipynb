```python
import numpy as np

arr1 = np.random.uniform(low=2, high=5, size=(5,5))
arr2 = np.random.rand(5,5)

print("Hola")
print(arr1)
print("--Arreglo 2--")
print(arr2)

vecX = []
vecY = []
vec_suma = []

for i in range(1):
    vecX.append(arr1[i])
    vecY.append(arr2[1])
    
print("--Vector X--")
print(vecX)
print("--Vector Y--")
print(vecY)
print("--Suma--")
for i in range(1):
    vec_suma.append(vecX[i] + vecY[i])
print(vec_suma)
```
