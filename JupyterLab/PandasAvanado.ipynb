```python
# Filtrar con una condición: doc.loc[condicion, columnas]
# valor simbolo(<,>,!=..) valor
# df['city'] == "CDMX"
df.loc[df['city']=="CDMX",['order_id','product','order_date','payment_method',"city"]]
```

```python
#Filtrar con varias condiciones
#'category','discount',
#Filtrar cuando categoría sea igual a electronico y tenga un descuento mayor a .10 = 10%
#Crear las condiciones en una variable
# , ['order_id','product','discount','payment_method','category']
#x = (df['category'] == ['Electronics']) & (df['discount']>0.1) 
#mask = (df['category'] == 'Electronics') & (df['discount'] > 0.1)
#df[mask] = df[['order_id', 'product', 'discount', 'payment_method', 'category']]
x = (df['category'] == 'Electronics') & (df['discount'] > 0.1)
df.loc[x, ['order_id', 'product', 'discount', 'payment_method', 'category']]
```

### Query - Consultas
- Especificas
- Simple

```python
#Mostrar todos los registros de CDMX
df.query("city == 'CDMX'")
```

```python
#Consulta que muestre CDMX, Electronicos, Descuento de .10
df.query("city=='CDMX' and category=='Electronics' and discount == .1")
```

```python
#mostrar los registros que sean de CMDMX, GDL, con un descuento igual a .1
dis = .1
ciudades = ["CDMX","GDL"]
df.query("discount == @dis and city in @ciudades")
#Busqueda por muchos valores, in hace referencia a varios.
```

```python
## Separando - Creando una nueva tabla
```

```python
users = (
    df[["user_id","user_name","segment"]]
    .drop_duplicates("user_id")
)
users.head()
```

```python
#merge, buscar el metodo para insertar "lef,right", on significa cual es el campo o la columna que coincide con ambas, en este caso
#el data frame original con el creado comparten user_id, many_to_many cardinalidad
df1 = df.merge(users, on="user_id",how="left",suffixes=("","_user"), validate="many_to_one")
df1.head()
#Se agrego la columna de user_name_user debido a que se identifica cuales se comparten
```
